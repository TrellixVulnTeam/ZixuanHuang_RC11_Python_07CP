```python
import networkx as nx
import csv
```

```python
with open("melb_data.csv",encoding='utf-8-sig')as csvfile:
    reader=csv.reader(csvfile)
    print(next(reader))
```

```python
class HousingPrice:
    def __init__(self,address,price,distance,room,lattitude,longtitude):
        self.address=address
        self.price=price
        self.distance=distance
        self.room=room
        self.lattitude=lattitude
        self.longtitude=longtitude
```

```python
with open("melb_data.csv")as csvfile:
    reader=csv.reader(csvfile)
    rows = 0
    for r in reader:
        rows +=1 
    print(rows)
```

```python
HousingPrices=[]
priceL=[]
with open("melb_data.csv")as csvfile:
    reader=csv.DictReader(csvfile)
    for row in reader:
        address = row['Address']
        price = row['Price']
        distance = row['Distance']
        room = row['Rooms']
        lattitude = row['Lattitude']
        longtitude = row['Longtitude']
        HousingPrices.append(HousingPrice(address,price,distance,room,lattitude,longtitude))
        
```

```python
import random
```

```python
priceTotal2=0
for h in HousingPrices:
    priceTotal2=priceTotal2+float(h.price)/float(h.room)


priceTotal2=priceTotal2/13580
print(priceTotal2)
```

```python
HousingPriceGraph = nx.Graph()
```

```python
HousingPricessubset=[]
random.shuffle(HousingPrices)
HousingPricessubset=HousingPrices[1:4000]
```

```python
i=0
for h in HousingPricessubset:
    n=float(h.distance)
    m=float(h.price)/float(h.room)
    HousingPriceGraph.add_node(h,pos=(n,m))
    i+=1
print(i)    
```

```python
options = {}
plt.figure(1, figsize = (30, 30))
pos=nx.get_node_attributes(HousingPriceGraph,'pos')
nx.draw(HousingPriceGraph,pos,node_color= 'black', node_size=20, width=1)
```
